### BMI Calculator
* The body mass index (BMI) is a measure used in medicine to see if someone is underweight or overweight. This is the formula used to calculate it bmi is equal to the __person's weight (in Kgs) divided by the person's height (in meters) squared__


```python
height = 1.65 
weight = 84

bmi = 84 / (1.65**2)   # weight in kgs / sq(height in meters) 
print(bmi)             # 30.85399449035813      
print(round(bmi))      # 31
print(round(bmi, 2))   # up to 2 decimal places 
```

### PROJECT: Tip calculator 

```python
print("Welcome to the tip calculator!")

bill = float(input("what was the total bill? $"))

tip = int(input("what percentage tip you would like to give? 10, 12 or 15 "))

count = int(input("How many people to split the bill?"))

total_bill = (tip/100) * bill + bill   

bill_per_person = round(total_bill/count,2)

print(f"Each person should pay: {bill_per_person}")
```
