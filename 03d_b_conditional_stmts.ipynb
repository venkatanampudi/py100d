### Nested if / else 
```

    if another condition:

    else:

```

````python
print("Welcome to the rollercoaster!")
height=int(input("What is your height in cm? "))

if height >= 120:
    print("You can ride the rollercoaster.")
    
    age = int(input("What is your age? "))
    if age <=12:
        print("Please pay $5.00")
    elif age <=18:   # this condition for age between 12 to 18
        print("Please pay $7.00")
    else:
        print("Please pay $12.00")
        
else:
    print("Sorry, you have to grow taller before you can ride.")
````

### ***Problem Algorithm***
* 1. Check the height so can ride?
* 2. if can ride, check the age for ticket bucketing 
* 3. after 1 & 2 are good, add +$3.00 to their ticket price if they want picture of their ride
* 4. else condition if none of these conditions met 




````python
print("Welcome to the rollercoaster!")
height=int(input("What is your height in cm? "))
bill = 0 

if height >= 120:
    print("You can ride the rollercoaster.")
    age = int(input("What is your age? "))
    
    if age <=12:
        bill = 5
        print("Child tickets are $5.00")
    elif age <=18: 
        bill = 7
        print("Youth tickets are $7.00")
    else:
        bill = 12
        print("Adults tickets are $12.00")
    wants_photo = input("Do you want to have a photo take? Type y for Yes and n for No.")
    if wants_photo == "y" or "Y":
        # Add $3 to their bill
        bill += 3
    print(f"Your final bill is ${bill}")
    
else:
    print("Sorry, you have to grow taller before you can ride.")
````

## Write a Python Pizza Delivery program 

````python
### Requirements 
'''
small pizza = $15
medium pizza = $20
large pizza = $25

Add pepparoni for small pizza (Y or N): +$2
Add pepparoni for medium or large pizza (Y or N): +$3
Add extra cheese for any size pizza (Y or N): +$1

Calculate the final bill with pizza & with customer add on choices.
'''
# Assignments 
size = input("What size pizza do you want? S, M or L: ")
pepparoni = input("Do you want pepparoni on your pizza? Y or N: ")
extra_cheese = input("Do you want extra cheese? Y or N")

# Init the bill amount first
bill = 0 

# assign the pizza price for each size 
if size == "S":
    bill += 15
elif size == "M":
    bill += 20
elif size == "L":
    bill += 25
else: 
    print("You typed the wrong inputs")

# How much to add to their bill based on their pepparoni choice
if pepparoni == "Y":
    if size == "S":
        bill += 2
    else:
        bill += 3

# How much to add to their final amount based on whether if they want extra cheese.
if extra_cheese == "Y":
    bill += 1

print(f"Your final amount is: ${bill}.00")

````
