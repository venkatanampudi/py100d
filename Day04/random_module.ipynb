### Creating random floating point numbers 
##### SY: random.random() --> Return the next random floating point number in the range of 0.0 <= x < 1.0  (<= can be greater or equal to 0)

```python
# first one is random module and second one is random function from random module
# Result is between 0.0 to 1.0 but not inclusive 1.0

random_num_0_to_1=random.random() 
print(random_num_0_to_1)
```

```python
# Multiply the same with 10 to check if randomisation between 1 & 10 but not inclusive 10
random_num_0_to_1=random.random() * 10
```
