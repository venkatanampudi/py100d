### Coding challenge - Banker Roulette. Who will pay the bill?
##### using choice() function of random


```python
import random
# Define the list of friends 
friends=["Alice","Bob","Charlie","David","Edward"]

# Option 1: use choice() function
print(random.choice(friends))

# Option 2: use index from random.randint(0,4) (both 0 & 4 included)
random_index = random.randint(0,4)
print(friends[random_index])
```

### Nested List (list with-in a list)

```python
fruits=["Strawberries","Apples","Grapes","Cherries","Pear","Peaches"]
veggies=["Spinach","Kale","Tomatoes","Potatoes","Onion"]

dirty_dozen=[fruits,veggies]

print(dirty_dozen)
print(dirty_dozen[1][1])
```

```python
fruits = ["Strawberries", "Nectarines", "Apples", "Grapes", "Peaches", "Cherries", "Pears"]
fruits[-1] = "Melons"
fruits.append("Lemons")
print(fruits)
```
